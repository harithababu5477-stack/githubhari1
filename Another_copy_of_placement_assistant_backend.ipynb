```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
!pip install sentence-transformers
```

```python
user_query = "What are ACID properties in DBMS?"
```

```python
query_embedding = embeddings.embed_query(user_query)
print(len(query_embedding))
```

```python
from langchain_community.vectorstores import Chroma

# load the existing DB
vector_db = Chroma(
    persist_directory="/content/placement_vectordb",
    embedding_function=embeddings
)

# retrieve top 3 similar chunks
results = vector_db.similarity_search_with_score(user_query, k=3)

for chunk, score in results:
    print(f"Score: {score}, Text: {chunk.page_content}")
```

```python
retriever = vector_db.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 4, "fetch_k": 10}
)

```

```python

results = vector_db.similarity_search("Explain ML", k=4)

for doc in results:
    print(doc.page_content)
```

```python
%%writefile app.py
import streamlit as st
import requests

st.set_page_config(page_title="AI & DS Placement Assistant", layout="wide")

st.title("🎓 AI & DS Placement RAG Chatbot")
st.write("Ask me anything related to AI & DS placements!")

# Backend API URL (Change this to your ngrok backend URL)
BACKEND_URL = "http://127.0.0.1:8000/ask"

user_question = st.text_input("Enter your question:")

if st.button("Ask"):
    if user_question:
        response = requests.post(
            BACKEND_URL,
            json={"query": user_question}
        )

        if response.status_code == 200:
            answer = response.json()["answer"]
            st.success(answer)
        else:
            st.error("Error from backend!")
```
